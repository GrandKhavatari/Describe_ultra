```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def monitoring(data, columns):
    intro = 0
    outro = 0
    for column in columns:
        intro = data[column].describe()
        intro['len'] = len(data[column])
        intro['NaN'] = data.loc[data[column].isna(), data.columns[0]].count()
        intro['NaN%'] = intro['NaN'] / intro['len'] * 100
        intro['-----'] = '-----'
        if data[column].dtype != object and data[column].dtype != bool:
            print(f"For \'{column}\' column monitor:")
            intro['delta'] = intro['max'] - intro['min']
            intro['median'] = data.loc[data[column].notna(), column].median()
            intro['outliers'] = data.loc[(
                                         (data[column] < (intro['25%'] - 1.5*(intro['75%'] - intro['25%']))) |
                                         (data[column] > (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))
                                        ), column].count() 
            
            intro['outliers %'] = intro['outliers'] / intro['len']*100 
            
            intro['norm_data_min'] = data.loc[(
                                         (data[column] > (intro['25%'] - 1.5*(intro['75%'] - intro['25%']))) &
                                         (data[column] < (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))
                                        ), column].min()
            intro['norm_data_max'] = data.loc[(
                                         (data[column] > (intro['25%'] - 1.5*(intro['75%'] - intro['25%']))) &
                                         (data[column] < (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))
                                        ), column].max()
              
            intro['down_outliers'] = data.loc[(
                                         (data[column] < (intro['25%'] - 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].count()
            
            intro['down_outliers_min'] = data.loc[(
                                         (data[column] < (intro['25%'] - 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].min()
            intro['down_outliers_max'] = data.loc[(
                                         (data[column] < (intro['25%'] - 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].max()
            
            intro['up_outliers'] = data.loc[(
                                         (data[column] > (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].count()
            intro['up_outliers_min'] = data.loc[(
                                         (data[column] > (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].min()
            intro['up_outliers_max'] = data.loc[(
                                         (data[column] > (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))                                         
                                        ), column].max()

            outro = intro[['-----',
                           'len',
                           'count',
                           'NaN',
                           'NaN%',
                           '-----',
                           'norm_data_max',
                           'norm_data_min',
                           'outliers',
                           'outliers %',
                           'up_outliers',
                           'up_outliers_max',
                           'up_outliers_min',
                           'down_outliers',
                           'down_outliers_max',
                           'down_outliers_min',
                           '-----',
                           'std', 
                           'mean', 
                           'median', 
                           '-----',
                           'max',
                           'min',
                           'delta',
                           '25%',
                           '50%',
                           '75%'
                          ]]

            outro['len'] = round(outro['len'])
            outro['count'] = round(outro['count'])
            outro['std'] = round(outro['std'],2)
            outro['mean'] = round(outro['mean'],2)
            outro['median'] = round(outro['median'],2)
            outro['outliers %'] = round(outro['outliers %'],2)
            outro['NaN'] = round(outro['NaN'])
            outro['NaN%'] = round(outro['NaN%'], 2)
            
            #print(a)
            print(outro)
            print()
            range_start = data[column].min()*0.9
            range_end = data[column].max()*1.1

            sns.set_theme(style='whitegrid')
            plt.figure(figsize=(20,0.8))
            plt.title(f'Data distribution "{column}"', fontsize=14)       
            boxplot = sns.boxplot(x=data[column], palette='Set3')
            boxplot.set_xlabel(None)
            boxplot.set_xlim((range_start), range_end)
            xlabels = ['{:,.2f}'.format(x) for x in boxplot.get_xticks()]


            sns.set_theme(style='whitegrid')
            plt.figure(figsize=(20,0.8))
            plt.title(f'Outliers distribution "{column}"', fontsize=14)
            boxplot = sns.boxplot(x=data.loc[((data[column] < (intro['25%'] - 1.5*(intro['75%'] - intro['25%']))) | (data[column] > (intro['75%'] + 1.5*(intro['75%'] - intro['25%'])))), column], palette='Set1')
            boxplot.set_xlabel(None)
            boxplot.set_xlim((range_start), range_end)
            xlabels = ['{:,.2f}'.format(x) for x in boxplot.get_xticks()]
            
        else:
            print(f"For \'{column}\' column monitor:")
            outro = intro[['-----',
                   'len',
                   'count',
                   'NaN',
                   'NaN%',
                   '-----',
                   'top', 
                   '-----',
                   'unique',
                   ]]
            outro['NaN'] = round(outro['NaN'])
            outro['NaN%'] = round(outro['NaN%'], 2)
            print(outro)        
```
